```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# creating a pie chart with a sample data
```

```python
labels = 'python', 'c++', 'Ruby', 'Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral', 'lightskyblue']
explode = (0.1,0,0,0) #exploding the first slice.
```

```python
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True,startangle=140)

plt.axis('equal')

plt.show
```

```python
# Plotting another set of pie chart
```

```python
labels = ['Cookies','Jellybean','Milkshake','Cheesecake']
sizes = [38.4,40.6,20.7,10.3]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']

patches, texts = plt.pie(sizes, colors = colors, shadow = True, startangle = 90)
plt.legend(patches, labels, loc="best")

plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
labels = ['Vennila','Strawberry','Butterscotch','Plain']
sizes = [38.4,40.6,20.7,10.3]
colors = ['green','yellow','red','blue']
explode = (1.6,0,0,0)

patches, texts = plt.pie(sizes, colors = colors, shadow = True, startangle = 130)
plt.legend(patches, labels)
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%')
plt.axis('equal')
plt.title('Range of Flavours')
plt.tight_layout()
plt.show()
```
